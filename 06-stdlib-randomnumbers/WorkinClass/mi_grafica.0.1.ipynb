```python
import pandas as pd
import matplotlib.pyplot as plt

def graficar(datos):

    path = "./"+ datos + ".txt"
    #data = pd.read_csv(path,sep=' ',names=['N','Delta'])
    data = pd.read_csv(path, delimiter='\s+', header=None, names=['x', 'y'], skiprows=1)


    plt.plot(data['x'],data['y'],"-",label='Delta')
    plt.title('Error relativo en funci√≥n de N')
    # plt.xscale("log")
    #plt.yscale("log")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.savefig('mi_grafica.pdf')
    plt.grid()
    plt.legend()
```

```python
graficar("data")
```
